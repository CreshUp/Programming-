```python
class Product:
    def __init__(self, name, price, category):
      self.name=name
      self.price=price
      self.category=category
      self.amount=10
    def is_available(self):
      if self.amount:
        return True
      else:
        return False

    def delete(self,how_much):
      self.amount-=how_much

    def add(self,how_much):
      self.amount+=how_much


class Order:
    def __init__(self,id,customer_id):
      self.id=id
      self.customer_id=customer_id
      self.products={}

    def price(self):
      price=0
      for v in self.products.values():
        price+=v[1]
      return price*1.25

    def add_to_cart(self,what,price):
      try:
        self.products[what][1]+=price
        self.products[what][0]+=1
      except Exception:
        self.products[what]=[0,0]
        self.products[what][1]+=price
        self.products[what][0]+=1

    def del_from_cart(self,what,price):
      self.products[what][1]-=price
      self.products[what][0]-=1

    def check(self):
      for k,v in self.products.items():
        if v[0]==0:
          continue
        print(f'item:{k} / amount:{v[0]} / price:{v[1]}')


class Customer:
    def __init__(self,customer_id):
      self.id=customer_id
      self.history={}

    def purchase(self, order):
      self.history[order.id]=order.price()
      print('You bought something with total price with НДС of', order.price())

# Немного не понял что должна была делать customer_cart, если я всё написал в order


products=[Product('Pomidor',100,'овощ'),
          Product('Chocolate',200,'сладкое')]
ord_id=0
customers=[]
while True:
  print('Добро пожаловать')
  name=input('Input Your name:')
  print('0 is No, 1 is Yes')
  t=False
  for customer in customers:
    if name == customer.name:
      t=True
      break
  if not t:
    customers.append(Customer(name))
  response=int(input('You want to buy some things: '))
  if response:
    order=Order(ord_id,name)
    ord_id+=1
    while True:
      response=int(input('You want to add something '))
      if response:
        for product in products:
          if product.is_available():
            response=int(input(f'We have {product.amount} of {product.name} with price {product.price} / Are u want to buy 1 of it '))
            if response:
              order.add_to_cart(product.name,product.price)
              product.delete(1)
          else:
            print(f'We dont have {product.name}')
          print('Current price with НДС is',order.price())
      else:
        break
      response=int(input('You want to delete something '))
      if response:
        for k,v in order.products.items():
          response=int(input(f'You have {v[0]} of {k} with price {v[1]} / Are u want to delete 1 of it '))
          if response:
            order.del_from_cart(k,v[1]/v[0])
            product.add(1)

  order.check()
  for customer in customers:
    if name==customer.id:
      customer.purchase(order)
      del order
      break
```

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
import random as r


class SpaceShip:
	dur=0
	fuel=0
	speed=0

	def __init__(self, durability, fuel):
		self.max_dur=durability
		self.max_fuel=fuel
		self.fuel+=fuel
		self.dur+=durability


	def is_broken(self,eng):
		if self.dur<=-100:
			print('Your ship cant be saved his dur is below -100')
			return eng, 0
		elif self.dur<=0:
			if not eng.busy and not eng.dead:
				print('You have a chance to save our ship...')
				print('but your Eng would hurt')
				ans=input('yes or else: ')
				if 'yes'==ans:
					self.dur+=100
					eng.health-=20
					eng.is_dead()
					eng.busy+=eng.task_speed
					eng.skill+=1
					print('Your Eng risk his life to save our ship')
					print(f"Ship dur is {self.dur}, Eng health is {eng.health}")

					return eng, 1
				else:
					print('Our ship has taken permanent damage')
					print('Max dur is reduced by 20')
					print('By the way, you still have to repair it')
					self.max_dur-=20
					if self.max_dur<100:
						print('Ship max dur cannt be bellow 100')
						print('You lost')
						return eng, 0
					else:
						return eng, 1
		else:
			return eng, 1


	def pilot_x_asteroid(self,pilot):
		if not pilot.busy:
			print('Your pilot isnt busy and would try to guide the ship through')
			ans=input('Should he do it, yes or else: ')
			if 'yes'==ans:
				if self.speed<=pilot.skill:
					print('Successful')
				else:
					print("Pilot's skill is too low")
					print("Unsuccessful, but pilot increased his skill")
					print('Ship dur -40')
					self.dur-=40
					pilot.skill+=1
					return pilot
			else:
				print('Pilot wouldnt complete the task')
				print('Ship dur -40')
				self.dur-=40
				return pilot

		else:
			print('Pilot cannt complete the task')
			print('Ship dur -40')
			self.dur-=40
			return pilot


	def speed_up(self,pilot):
		if not pilot.busy:
			print('Pilot can speed up the Ship, should he?')
			ans=input('yes or else: ')
			if 'yes'==ans:
				print(f'Current speed is {self.speed}')
				try:
					ans=int(input('How much should he add: '))
					if ans<=0:
						print('Thats not how it works')
						print('Nothing has happened')
						return pilot
					else:
						self.speed+=ans
						pilot.busy+=pilot.task_speed
						print(f'Pilot will be busy while pathing route with new speed for {pilot.task_speed} turns')
						return pilot
				except Exception:
					print('Thats not how it works')
					print('Nothing has happened')
					return pilot
			else:
				return pilot
		else:
			print('Pilot is still busy for',pilot.busy,'turns')
			return pilot


	def repair(self, eng):
		if not eng.busy:
			print('Eng isnt busy so he can repair the ship')
			if self.dur==self.max_dur:
				print('Eng found out there is nothing to repair')
				return eng
			else:
				ans=input('Should he (yes or else): ')
				if ans=='yes':
					d1=self.dur
					self.dur+=eng.skill
					dif=self.dur-self.max_dur
					if dif>0:
						self.dur-=dif
					eng.busy+=eng.task_speed
					print(f"Eng is repairing {self.dur-d1} dur, he'll be busy for {eng.task_speed} turns")
					return eng
		else:
			print('Eng is busy for',eng.busy,'turns')
			return eng
		return eng


	def slow_down(self,pilot):
		if not pilot.busy:
			print('Pilot can slow down the Ship, should he?')
			ans=input('yes or else: ')
			if 'yes'==ans:
				print(f'Current speed is {self.speed}')
				try:
					ans=int(input('How much should he unadd: '))
					if ans<=0:
						print('Thats not how it works')
						print('Nothing has happened')
						return pilot
					else:
						self.speed-=ans
						if self.speed<0:
							self.speed=0
						pilot.busy+=pilot.task_speed
						print(f'Pilot will be busy while pathing route with new speed for {pilot.task_speed} turns')
						return pilot
				except Exception:
					print('Thats not how it works')
					print('Nothing has happened')
					return pilot
		else:
			print('Pilot is still busy for',pilot.busy,'turns')
			return pilot
		return pilot


	def fuel_adding(self,eng):
		if not eng.busy:
			print('Eng can add fuel, should he?')
			ans=input('yes or else: ')
			if ans=='yes':
				self.fuel+=eng.skill*2
				eng.skill+=2
				print(f'Eng added {eng.skill*2} of fuel, he will be busy for {eng.task_speed} turns')
				print('His skill are growing by 2')
				return eng
			else:
				print('Nothing had happend')
				return eng
		else:
			print('Eng is too busy to add fuel')
			return eng


class Mission:
	destination=1000  #Единиц расстояния до мнимой цели :)


	def is_asteroid(self,ship,pilot,eng):
		if r.randrange(1,11)<3:
			print('Asteroid on the way!')
			if ship.speed>100:
				print('Your speed is too high, -50 durability')
				ship.dur-=50
				eng,mission_status=ship.is_broken(eng)
			elif ship.speed==0:
				print('Punch straight in da face, -20 durability')
				ship.dur-=20
				eng,mission_status=ship.is_broken(eng)
			elif ship.speed<=100:
				pilot=ship.pilot_x_asteroid(pilot)
				eng,mission_status=ship.is_broken(eng)
		else:
			print('Everything is suspiciously still... too still')
		return ship,pilot,eng


	def movement(self,ship):
		if ship.speed<=ship.fuel:
			ship.fuel-=ship.speed
			self.destination-=ship.speed
			print(f"{ship.speed} of fuel is consumed, {ship.fuel} left")
		else:
			self.destination-=ship.fuel
			print('The fuel run out, so the Ship travelled only for', ship.fuel)
			print('Speed is 0 also')
			ship.fuel=0
			ship.speed=0
		return ship


	def is_arrived(self):
		if self.destination<=0:
			return True
		else:
			return False


class CrewMember:
	health=100
	busy=0
	task_speed=1 #Сколко ходов выполняют задачу
	dead=False


class Pilot(CrewMember):

	def __init__(self, skill):
		self.skill=skill


	def is_dead(self):
		if self.health<1:
			self.dead=True
			print('Pilot is dead')


class Enginier(CrewMember):

	def __init__(self, skill):
		self.skill=skill


	def is_dead(self):
		if self.health<1:
			self.dead=True
			print('Eng is dead')
		else:
			self.dead=False


pilot=Pilot(50)
enginier=Enginier(50)
mission=Mission()
spaceship=SpaceShip(1000,2000)
print(f"Welcome to Spaceship simulator, you have to travel {mission.destination} or else...")

while True:
	if not pilot.dead:
		pilot=spaceship.speed_up(pilot)
		pilot=spaceship.slow_down(pilot)
	else:
		print('Pilot is dead')

	if not enginier.dead:
		enginier=spaceship.fuel_adding(enginier)
		enginier=spaceship.repair(enginier)
	else:
		print('Eng is dead')

	spaceship,pilot,enginier=mission.is_asteroid(spaceship,pilot,enginier)
	spaceship=mission.movement(spaceship)
	if mission.is_arrived():
		print('YOU WIN')
		break
	enginier, status=spaceship.is_broken(enginier)
	if not status:
		print('YOU LOST')
		break
	if pilot.busy:
		pilot.busy-=1
	if enginier.busy:
		enginier.busy-=1
	print(f"{mission.destination} has left")



```
